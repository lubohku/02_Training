```python
import  xml.dom.minidom
import pandas as pd

dom = xml.dom.minidom.parse('mask_lyp.lyp')
df = pd.DataFrame(columns=['source', 'frame-color', 'fill-color', 'dither-pattern', 'visible', 'transparent', 'width', 'animation'])
#得到文档元素对象
root = dom.documentElement

lyp_idx = []
for tag in df.columns:
    lyps = root.getElementsByTagName(tag)
    temp_lyp = []
    for lyp in lyps:
        if tag == "source":
            lyp_data = lyp.firstChild.data.split("@")[0].split(" ")[0]
            lyp_idx.append(lyp.firstChild.data.split("@")[0].split(" ")[1])
        elif tag == 'frame-color' or tag == 'fill-color':
            lyp_data = lyp.firstChild.data[1:]
        else:
            lyp_data = lyp.firstChild.data
        temp_lyp.append(lyp_data)
    df[tag] = temp_lyp
# print(df['source'])
# print(lyp_idx)
df.insert(1, 'layer-index', lyp_idx, allow_duplicates=False)
# print(df['Layer Index'])
df.columns = ['Name', 'Layer', 'Frame', 'Fill', 'Dither', 'Visible', 'Transparent', 'Width', 'Animation']
df.to_excel('mask_lyp.xlsx', index = False)
df.head()
```
